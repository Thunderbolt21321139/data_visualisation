```python
# count plot examples

import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
# prepare the data

print(len(df))
df = df[df['cylinders'].isin([4,6,8])].copy()
print(len(df))

df['cylinders'] = df['cylinders'].astype(str) + ' cylinders'

df.head()
```

```python
# basic box plot

sns.catplot(x='origin', y='horsepower', data=df, kind='box', hue='cylinders')
plt.show()
```

```python
# remove outliers

sns.catplot(x='origin', y='horsepower', data=df, kind='box', hue='cylinders', flierprops={'marker': ''})
plt.show()
```

```python
# adjust whiskers

sns.catplot(x='origin', y='horsepower', data=df, kind='box', hue='cylinders', whis=[0,100])
plt.show()
```
