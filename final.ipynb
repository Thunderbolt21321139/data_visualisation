```python
# count plot examples

import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
# prepare the data

df['cylinders'] = df['cylinders'].astype(str) + ' Cylinders'
df['origin'] = df['origin'].str.capitalize()
df['origin'] = df['origin'].replace('Usa', 'America')
df.head()
```

```python
# display bar chart of horsepower per cylinder for each origin

sns.set_style('white')

order = ['3','4','5','6','8']
order = [cyl + ' Cylinders' for cyl in order]

palette = sns.color_palette(['#003049','#d62828','#f77f00'])

g = sns.catplot(data=df, kind='bar', x='origin', y='horsepower', col='cylinders', hue='origin', palette=palette, legend='full')
g.figure.suptitle("Average Horsepower by Cylinders and Origin", fontsize=18, y=1.05, fontweight='bold')

# add x y labels
g.set(xlabel="", ylabel="Horsepower (hp)")

# remove x tick labels
g.set_xticklabels("")

for ax in g.axes.flatten():
    cyl = ax.get_title().split(' = ')[-1]  # Get the origin from the title
    ax.set_title(f'{cyl}', fontsize=12)  # Set a new subtitle

plt.show()
```
