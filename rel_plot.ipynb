```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
df['origin'].value_counts()
```

```python
df.head()
```

```python
col_order = [8,6,5,4,3]
row_order = ['europe','japan','usa']

sns.relplot(x='horsepower', y='acceleration', data=df, kind='scatter', col='cylinders', col_order=col_order, row='origin', row_order=row_order)
plt.show()
```
