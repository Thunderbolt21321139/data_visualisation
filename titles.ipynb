```python
# count plot examples

import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
# prepare the data

df['cylinders'] = df['cylinders'].astype(str) + ' cylinders'
df.head()
```

```python
# facet grid is used to make subplots from variable, axessubplot is used to make more customised subplots

g = sns.relplot(data=df, kind="scatter", x="weight", y="horsepower")
print(type(g))

g = sns.lineplot(data=df, x="model_year", y="horsepower")
print(type(g))
```

```python
# Add a title to facetgrid

sns.relplot(data=df, kind="scatter", x="weight", y="horsepower")
plt.title("Car Weight vs. Horsepower")
plt.show()
```

```python
# Add a title to axessubplot

g = sns.lineplot(data=df, x="model_year", y="horsepower")
plt.title("Average Horsepower Over Time")
plt.show()
```

```python
# Add x and y labels

sns.relplot(data=df, kind="scatter", x="weight", y="horsepower")
plt.title("Car Weight vs. Horsepower")
plt.xlabel("Weight")
plt.ylabel("Horsepower")
plt.show()
```

```python
# rotate x labels

sns.relplot(data=df, kind="scatter", x="weight", y="horsepower")
plt.title("Car Weight vs. Horsepower")
plt.xlabel("Weight")
plt.xticks(rotation=90)
plt.ylabel("Horsepower")
plt.show()
```
