```python
# count plot examples

import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
# prepare the data

df.head()
```

```python
# basic point plot

sns.catplot(x='cylinders', y='horsepower', data=df, kind='point')
plt.show()
```

```python
# add caps

sns.catplot(x='cylinders', y='horsepower', data=df, kind='point', capsize=0.2)
plt.show()
```

```python
# remove lines

sns.catplot(x='cylinders', y='horsepower', data=df, kind='point', capsize=0.2, linestyle='none')
plt.show()
```

```python
# multiple lines

sns.catplot(data=df, kind='point', x='cylinders', y='horsepower', hue='origin')
plt.show()
```

```python
# remove confidence intervals

sns.catplot(data=df, kind='point', x='cylinders', y='horsepower', hue='origin', errorbar=None)
plt.show()
```

```python
# use custom metric

from numpy import median

sns.catplot(data=df, kind='point', x='cylinders', y='horsepower', hue='origin', errorbar=None, estimator=median)
plt.show()
```
