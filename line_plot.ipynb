```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
col_order = [8,6,5,4,3]
row_order = ['europe','japan','usa']

sns.relplot(x='model_year', y='mpg', data=df, kind='line', col='cylinders', col_order=col_order, row='origin', row_order=row_order, errorbar='sd')
plt.show()
```

```python
sns.relplot(x="model_year", y="horsepower", data=df, kind="line", errorbar=None, style="origin", hue="origin", markers=True, dashes=False)
plt.show()
```
