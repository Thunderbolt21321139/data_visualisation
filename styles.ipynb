```python
# count plot examples

import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('mpg')
```

```python
# prepare the data

df['cylinders'] = df['cylinders'].astype(str) + ' cylinders'
df.head()
```

```python
# use whitegrid style

sns.set_style('whitegrid')

order = ['3','4','5','6','8']
order = [cyl + ' cylinders' for cyl in order]

sns.catplot(data=df, kind='count', x='cylinders', order=order, hue='cylinders')
plt.show()
```

```python
# use purples palette

sns.set_style('whitegrid')

order = ['3','4','5','6','8']
order = [cyl + ' cylinders' for cyl in order]
palette = sns.color_palette('Purples', n_colors=len(order))
new_palette = [None] * 5
new_palette[0] = palette[4]
new_palette[1] = palette[1]
new_palette[2] = palette[3]
new_palette[3] = palette[0]
new_palette[4] = palette[2]

sns.catplot(data=df, kind='count', x='cylinders', hue='cylinders', order=order, palette=new_palette)
plt.show()
```

```python
# use diverging red blue palette

sns.set_style('whitegrid')

order = ['3','4','5','6','8']
order = [cyl + ' cylinders' for cyl in order]

palette = sns.color_palette('RdBu', n_colors=len(order))
new_palette = [None] * 5
new_palette[0] = palette[4]
new_palette[1] = palette[1]
new_palette[2] = palette[3]
new_palette[3] = palette[0]
new_palette[4] = palette[2]

sns.catplot(data=df, kind='count', x='cylinders', hue='cylinders', order=order, palette=new_palette)
plt.show()
```

```python
# change scale to paper

sns.set_style('whitegrid')
sns.set_context('paper')

order = ['3','4','5','6','8']
order = [cyl + ' cylinders' for cyl in order]

palette = sns.color_palette('Blues', n_colors=len(order))
new_palette = [None] * 5
new_palette[0] = palette[4]
new_palette[1] = palette[1]
new_palette[2] = palette[3]
new_palette[3] = palette[0]
new_palette[4] = palette[2]

sns.catplot(data=df, kind='count', x='cylinders', hue='cylinders', order=order, palette=new_palette)
plt.show()
```

```python
# use a custom palette

sns.set_style('darkgrid')

order = ['3','4','5','6','8']
order = [cyl + ' cylinders' for cyl in order]

# https://coolors.co/a1cca5-2d93ad-613dc1-4e148c-14080e
palette = sns.color_palette(['#292E1E','#7F0799','#9649CB','#AFBBF2','#D3FFF3'])
new_palette = [None] * 5
new_palette[0] = palette[4]
new_palette[1] = palette[1]
new_palette[2] = palette[3]
new_palette[3] = palette[0]
new_palette[4] = palette[2]

sns.catplot(data=df, kind='count', x='cylinders', hue='cylinders', order=order, palette=new_palette)
plt.show()
```
