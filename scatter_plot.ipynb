```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('diamonds')
```

```python
df = df[df['clarity'] == 'I1']
```

```python
df.head()
```

```python
sns.scatterplot(x='carat', y='price', data=df, size='cut', style='cut', hue='cut', hue_order=['Fair','Good','Very Good','Premium','Ideal'])
plt.show()
```
